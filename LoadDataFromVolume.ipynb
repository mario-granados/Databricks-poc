```python
%fs ls /Volumes/catalog_mario_poc/bronze/sourcedata/addresses/
```

```python
# Read CSV files from Volume
df = spark.read.format("csv") \
    .option("delimiter", ",") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("/Volumes/catalog_mario_poc/bronze/sourcedata/addresses/")

display(df)
```

```python
df.write.mode("overwrite").saveAsTable("catalog_mario_poc.bronze.customer_addresses_raw")
```

```python
%sql select * from catalog_mario_poc.bronze.customer_addresses_raw
```
